```python
from sklearn.ensemble import VotingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
# Base Models
dt = DecisionTreeClassifier(random_state=42)
lr = LogisticRegression(max_iter=1000)
knn = KNeighborsClassifier()

# Voting Classifier
vote = VotingClassifier(
    estimators=[
        ('dt', dt),
        ('lr', lr),
        ('knn', knn)
    ],
    voting='hard'
)

# Train
vote.fit(X_train_cls, y_train_cls)

# Predict
y_pred_vote = vote.predict(X_test_cls)

print("Voting Accuracy:", accuracy_score(y_test_cls, y_pred_vote))
```

```python
# Hard Voting
hard = VotingClassifier(
    estimators=[('dt', dt), ('lr', lr), ('knn', knn)],
    voting='hard'
)

hard.fit(X_train_cls, y_train_cls)

# Soft Voting
soft = VotingClassifier(
    estimators=[('dt', dt), ('lr', lr), ('knn', knn)],
    voting='soft'
)

soft.fit(X_train_cls, y_train_cls)

print("Hard Voting :", accuracy_score(y_test_cls, hard.predict(X_test_cls)))
print("Soft Voting :", accuracy_score(y_test_cls, soft.predict(X_test_cls)))
```

Hard Voting selects the class with the most votes.

Soft Voting averages prediction probabilities and usually gives better accuracy.

```python
from sklearn.ensemble import StackingClassifier

stack_cls = StackingClassifier(
    estimators=[
        ('dt', DecisionTreeClassifier()),
        ('knn', KNeighborsClassifier())
    ],
    final_estimator=LogisticRegression(max_iter=1000)
)

stack_cls.fit(X_train_cls, y_train_cls)

y_pred_stack = stack_cls.predict(X_test_cls)

print("Stacking Accuracy:", accuracy_score(y_test_cls, y_pred_stack))
```

```python
from sklearn.ensemble import StackingRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import r2_score
```

```python
stack_reg = StackingRegressor(
    estimators=[
        ('dt', DecisionTreeRegressor()),
        ('knn', KNeighborsRegressor())
    ],
    final_estimator=LinearRegression()
)

stack_reg.fit(X_train_reg, y_train_reg)

y_pred_stack_reg = stack_reg.predict(X_test_reg)

print("Stacking R2 Score:", r2_score(y_test_reg, y_pred_stack_reg))
```
